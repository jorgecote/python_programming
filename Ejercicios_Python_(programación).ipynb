###Ejercicio 12 (Distancia entre 2 puntos en la tierra)
La superficie de la tierra es curva, y la distancia en grados de longitud varía con la latitud. Como resultado, encontrar la distancia en tre dos puntos en la tierra es más complicado que aplicar un teorema de pitagoras.

Suponga (t1, g1) y (t2, g2) sea la longitud y latitud de 2 puntos en la superficie de la tierra. La distancia promedoio entre dos puntos en la superficie de la tierra esta dada como:

d= 6371.01 arccos((sin(t1)sin(t2))+(cos(t1)cos(t2)cos(g1-g2))

Crae un programa que permita al usuario ingresar la longitud y latitud de dos puntos (en grados) en la superficie de la tierra y calcule la distancia entre dichos puntos en kilometros

```python
from math import acos
import math
t1 = float(input("Ingrese la longitud del punto 1: "))
g1 = float(input("Ingrese la latitud del punto 1: "))
t2 = float(input("Ingrese la longitud del punto 2: "))
g2 = float(input("Ingrese la latitud del punto 2: "))

t1=t1*180/math.pi
g1=g1*180/math.pi
t2=t2*180/math.pi
g2=g2*180/math.pi

d= 6371.01*acos((math.sin(t1)*math.sin(t2))+(math.cos(t1)*math.cos(t2)*math.cos(g1-g2)))
print(f"La distancia de los dos puntos en kilometros es: {d} Km")
```

### Ejercicio 23 (Área de un polígono regular)
Un polígono regular es aquel cuyos lados son todos iguales y sus los ángulos entre todos los lados adyacentes son todos iguales. El área de un polígono regular se puede calcular como la fórmula que sigue, sabiendo que s es la longitud de un lado y n el número de lados.

area = n * s^2 / (4 * tan(pi/2))

Escriba un programa que lea de teclado el tamaño de lado s y el número de lados de un polígono regular y muestre el área del mismo.

### Ejercicio 50 (Raíces de una función cuadrática)
Una función univariada cuadráticatiene la forma f(x) = ax^2 + bx + c, donde a, b y c con constantes, a es diferente de cero. Las raíces de una función cuadrática pueden ser encontradas, buscando las x que satisfacen la ecuación ax^2 + bx + c = 0. Una ecuación cuadrática puede tener 0, 1 o 2 raíces reales. Estas raíces pueden ser encontradas por medio de la fórmula cuadrática:

root = (-b +- sqrt(b^2 - 4ac))/(2a)

La parte que está dentro del radical se conoce como el discriminante. Si el discriminante negativo entonces la función no tiene raíces reales. Si el discriminante es 0 la ecuación tiene una raíz real. En cualquier otro caso tiene 2 raíces reales las cuales pueden ser evaluadas una con el signo  +  y la otra con el signo -.
Escriba un programa que calcule la cantidad de raíces reales de una función cuadrática. El programa debe utilizar los valores de a, b y c ingresados por el usuario y mostrar la cantidad de raíces reales y sus respectivos valores.

```python
import math
a = int(input("Ingrese el valor de a: "))
b = int(input("Ingrese el valor de b: "))
c = int(input("Ingrese el valor de c: "))

if ((b**2) - 4 * a * c) < 0:
  print("La función no tiene raíces reales")
elif ((b**2) - 4 * a * c) == 0:
  root = (-b + math.sqrt(b**2 - 4 * a * c))/(2 * a)
  print(f"La función tiene 1 raíz real, cuyo valor es {root}")
else:
  root1 = (-b + math.sqrt(b**2 - 4 * a * c))/(2 * a)
  root2 = (-b - math.sqrt(b**2 - 4 * a * c))/(2 * a)
  print(f"La función tiene 2 raíz real, cuyo valor es {root1} y {root2}")
```

La idea detras de este cifrado era muy simple (Con las técnicas de desencripción actuales no es una protección acptable). Cada letra en el mensaje original es desplazado 3 lugares. Como resultado, A se convierte en una D, la C se convierte en una F, la de en una G, etc. Las últimas 3 letras en el alfabeto son conectadas con el inicio del alfabeto, es decir X se convierte en A, Y se convierte en B y Z en C. *  inicialice supuesto como x/2
*  Mientras supuesto no es lo suficientemente bueno 
   
    actualice supuesto para ser el promedio de supuesto y x/supuesto

Cuando el algoritmo se completa, supuesto contiene una aproximación de la raíz cuadrada. La calidad de la aproximación depende de como se define "suficientemente bueno" .
```python
x = float(input("Ingrese el número: "))

supuesto = x / 2
while ((supuesto**2) - x) >= 10**(-12):
  supuesto = 1 / 2 * (supuesto + (x / supuesto))
print(f"La raíz cuadrada de {x} es {supuesto: .2f}")
```

```python
def gcd(n, m):          # Calcula el máximo común divisor entre 2 enteros
  if n < m: d = n
  else: d = m
  while m % d != 0 or n % d != 0:
    d -= 1
  return d


def reduce_frac(num, den):
  d = gcd(num, den)
  return num / d , den / d
```

Considere los strings kitten y sitting. La primera string puede ser transformda en la segunda con las siguientes operaciones: subustituya k por s, substituya e con una i e inserte g al final del string. Este es el mínimo número de operaciones que pueden ser realizadas para transformar kitten y sitting. Como resultado, la distancia edit es 3.

image.png

Use su función recursiva para escribir un programa que lea 2 string desde el usuario y muestre la distancia edit entre ellas.

```python
def edit_dist(s,t):
  if len(s) == 0:
    return len(t)
  elif len(t) == 0:
    return len(s)
  else:
    cost = 0
    if s[-1] != t[-1]:
      cost = 1
    d1=edit_dist(s[:-1],t)+1
    d2=edit_dist(s,t[:-1])+1
    d3=edit_dist(s[:-1],t[:-1])+cost
    return min(d1,d2,d3)

```

### Ejercicio 173 (Decodificación Run-Length)
La codificación Run-length es una simple técnica de compresión de datosque puede ser efectiva cuando valores repetidos ocurren en posiciones adyacentes dentro de una lista. La compresión es lograda reemplazando grupos de valores repetidos conuna copia del valor, seguido por el número de veces que se ha repetido. Por ejemplo, la lista ["A","A","A","A","A","A","A","A","A","A","A","A","B","B","B","B","A","A","A","A","A","A","B"] sería comprimido como ["A",12,"B",4,"A",6,"B",1]. La descompresión es llevada a cabo replicando el símbolo en la lista el número de veces indicado.

Escriba una función recursiva que descomprima una codificación Run-length. Su función recursiva tomará como parámetro una lista con codificación Run-length como único parámetro. Retornará como resultado una lista descomprimida como único resultado. Crear un programa principal que muestre una lista con codificación Run-length y el resultado de decodificarla.

```python
def RL_decode(rl_seq, rl_dec=[]):
  try:
    rl_dec += rl_seq[0]*rl_seq[1]
    rl_seq = rl_seq[2:]
    return RL_decode(rl_seq,rl_dec)
  except: return rl_dec
```

```python
rl_seq = ["A",12,"B",4,"A",6,"B",1]
print(RL_decode(rl_seq))
```
